## Variables

***So, what is a variable? Consider that your house needs a name. You place a nameplate at the front gate of your house. People will now recognize your house through that nameplate. That nameplate can be considered as variable. Like a nameplate points to the house, a variable points to the value that is stored in memory. When you create a variable, the interpreter will reserve some space in the memory to store values.***

```python
# Assign s as a string
s = "Hello Python"
s
```

```python
print(s)
```

### Set Operations using Methods and Builtins

```python
# Intersection 
#a.intersection(b) returns a new set with elements present in both a and b

a = {1, 2, 2, 3, 4}
b = {3, 3, 4, 4, 5}

a.intersection(b)
```

```python
#Union
#a.union(b) returns a new set with elements present in either a and b

a = {1, 2, 2, 3, 4}
b = {3, 3, 4, 4, 5}

a.union(b)
```

```python
#Difference
#a.difference(b) returns a new set with elements present in a but not in b

a.difference(b)
```

```python
#Symmetric Difference
#a.symmetric_difference(b) returns a new set with elements present in either a or b but not in both

a.symmetric_difference(b)
```

```python
#  Get the unique elements of a list

f1 = [3, 6, 2, 1, 3, 6, 7, 5, 2, 1, 5, 8, 9]

u = set(f1)
u
```
